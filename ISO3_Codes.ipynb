```python
import requests
import pandas as pd
from bs4 import BeautifulSoup  
from lxml import html
```

```python
html_response = requests.get('https://www.iban.com/country-codes')
html = html_response.text
html
```

```python
soup = BeautifulSoup(html)
```

```python
country = list(soup.select("#myTable > thead > tr")[0].children)[1].string
country
```

```python
iso_code = list(soup.select("#myTable > thead > tr")[0].children)[5].string
iso_code
```

```python
table = soup.select("#myTable > tbody > tr")
table
```

```python
country_codes = {}

for country_code in table:
    country_info: = {
        "country": country_code.select("td")[0].get_text(),
        "iso3_code": country_code.select("td")[2].get_text()
    }
    country_codes.append(country_info)

country_codes
```

```python
country_codes = {}

for country_code in table:
    country = country_code.select("td")[0].get_text()
    iso3_code = country_code.select("td")[2].get_text()

    country_codes[country] = iso3_code

country_codes
```

```python
code_dictionary = {k:v for row in country_codes for (k,v) in row.items()}
code_dictionary
```

```python
result = {}
for d in country_codes:
    result.update(d)
result
```
